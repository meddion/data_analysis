```python
from math import sin
import random
import matplotlib.pyplot as plt
import numpy as np

# To reproduce the plots
random.seed(42)

# Variant = 11
func = lambda x: sin(0.1*x)**2 + sin(x)**2

noisy_func = lambda x: func(x) + random.random()*0.5

X = list(np.arange(0, 10, 0.1))

Y = [noisy_func(i) for i in np.arange(0., 10., 0.1)]
plt.figure(figsize=(15, 10))
plt.plot(X, Y, color="red")

# Метод обновлюваної середньої
def avg_method(data):
    assert len(data) > 0
    points = [data[0]]
    n = 2
    return points+[points[-1]+1/n*(point - points[-1]) for point in data[1:]]

plt.plot(X, avg_method(Y), color="blue")

# Дослідити залежність від значення α в методі експоненційного згладжуванн
def exp_smoothing(data, α = 0.6):
    assert len(data) > 0
    ps = [data[0]]
    return ps + [ps[-1] + α * (p - ps[-1]) for p in data[1:]]

plt.plot(X, exp_smoothing(Y, α=0.99), color="green")

plt.show()
```
