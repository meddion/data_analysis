```python
from math import sin
import random
import matplotlib.pyplot as plt
import numpy as np

# To reproduce the plots
random.seed(42)
plt.figure(figsize=(15, 10))

# Variant = 11
f = lambda x: sin(0.1*x)**2 + sin(x)**2

x = np.linspace(0,10,100,endpoint=False)
y = []
for i in x:
    y.append(f(i)+ np.random.rand()/2)
plt.plot(x,y, color="blue")

alpha = 0.2
y_smooth = []
y_previous = y[0]

for i in range(0,len(y)):
    y_smooth.append(alpha * y[i] + (1 - alpha) * y_previous)
    y_previous = y_smooth[-1]
plt.plot(x, y_smooth, color = "red")

y = []
for i in x:
    y.append(f(i))
plt.plot(x,y, color="yellow")

y_smooth = []
y_previous = y[0]

for i in range(0,len(y)):
    y_smooth.append(y_previous + (y[i]-y_previous)/(i+1))
    y_previous = y_smooth[-1]

plt.plot(x, y_smooth, color="green")
```
