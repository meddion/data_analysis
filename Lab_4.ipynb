```python
from math import exp, cos
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Variant = 11
print("The bar diagram shows the dynamic of the urban population in the Kyiv region from 1989 to 2014.")

data = pd.read_csv("./population_data.csv").loc[0:25,:]
data.plot(x="year", y="population", kind="bar", figsize=(11, 6), title="The absolute values")

scaler = MinMaxScaler()
data["population"] = scaler.fit_transform(data["population"].to_numpy().reshape(-1, 1))
data.plot(x="year", y="population", kind="bar", figsize=(11, 6), title="The scaled version (0,1)")
plt.show()
```

```python
#city, village = pd.read_csv("./city_selo.csv").loc["city", "village"]
#village
labels = ["city", "village"]
x = pd.read_csv("./city_selo.csv").iloc[0][labels]
print(f"The year 2000:\nCity's population: {x[0]};\nVillage's population: {x[1]}.")
plt.figure(figsize=(10, 5))
plt.pie(x=x, labels=labels, autopct='%1.1f%%', startangle=90)
plt.show()
```

```python
my_group = [
"Роман",
"Захар",
"Роман",
"Ярослав",
"Володимир",
"Богдан",
"Роман",
"Віталій",
"Назарій",
"Володимир"
]
name_count = {}
for name in my_group:
    key = name[0].lower()
    name_count[key] = 1+name_count.get(key, 0)

plt.bar(list(name_count.keys()), list(name_count.values()))
plt.show()
```
